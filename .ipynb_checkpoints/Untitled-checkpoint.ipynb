```python

def bar_nanValue(x):                                                # Function to show missing values in the column to drop the column
    
    missing_percent = x.isnull().mean() * 100
    
    # Plotting the bar plot
    plt.figure(figsize=(15, 6))
    sns.barplot(x=missing_percent.index, y=missing_percent, color='skyblue')
    plt.axhline(y=40, color='r', linestyle='--')
    plt.title('Percentage of Missing Values by Column')
    plt.xticks(rotation=45, ha='right',fontsize=7)
    plt.xlabel('Columns')
    plt.ylabel('Percentage of Missing Values')
    plt.ylim(0, 100)  # Set limit to 100% for better visualization
    
    
    for i, value in enumerate(missing_percent):
        if value > 40:
            plt.text(i, value + 2, f'{value:.1f}%', ha='center', va='bottom', color='black')
    
    plt.show()

bar_nanValue(train)
```

```python
def drop_col(x):                                              # Function for Dropping column which has more than 40% missing values
    threshold = 0.4
    missing_percent= x.isnull().mean() * 100
    
    # Drop columns with missing values above threshold
    columns_to_drop = missing_percent[missing_percent > threshold * 100].index
    x = x.drop(columns=columns_to_drop,inplace= True)
    
    print(f'Dropped columns: {columns_to_drop}')
    print(train.shape)

drop_col(train)
```

```python

def handle_missing(df):
    import missingno as msno
    # Create a DataFrame to store the correlation of missing values
    cols_with_missing = df.columns[df.isnull().any()]
    missing_corr = df[cols_with_missing].isnull().corr()
    
    # Step 2: Increase the Size of the Plot
    plt.figure(figsize=(15, 12))
    
    # Step 3: Annotate Only Significant Correlations and Mask the Lower Triangle
    mask = np.triu(np.ones_like(missing_corr, dtype=bool))
    sns.heatmap(missing_corr, annot=True, mask=mask, cmap='coolwarm', vmin=-1, vmax=1, annot_kws={"size": 8})
    plt.title("Correlation Matrix of Missing Values")
    plt.show()
    
    # Step 4: Use Clustering to Group Similar Columns
    msno.dendrogram(df)
    plt.show()
    
    sns.clustermap(missing_corr, annot=True, cmap='coolwarm', figsize=(15, 12), annot_kws={"size": 8})
    plt.title("Clustered Correlation Matrix of Missing Values")
    plt.show()

handle_missing(train)
```

```python
# Example: Mean imputation for numerical columns
def missing_value(df):
    for col in df.select_dtypes(include='number').columns:
        df[col].fillna(df[col].mean(), inplace=True)
    
    # Example: Mode imputation for categorical columns
    for col in df.select_dtypes(include='object').columns:
        df[col].fillna(df[col].mode()[0], inplace=True)
```
